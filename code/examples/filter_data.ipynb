```python
from non_linear_beam.code.module.nonlinearprocess import get_frequency_data, get_time_data, set_file_name, plot_fft, bandpass_filter
import zipfile
import matplotlib.pyplot as plt
```

```python
dir = "/content/non_linear_beam/Experimental Data/"

time_mec_dir_t1 = set_file_name(dir, time = True, mec = True,test=1)
```

```python
time_mec_1, vel_mec_1 = get_time_data(time_mec_dir_t1)
```

```python
plt.figure(figsize = (10,7))
plt.plot(time_mec_1, vel_mec_1)

# Adding labels and title
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Time vs Amplitude')
plt.grid()

# Display the plot
plt.show();
```

```python
_,_ = plot_fft(time_mec_1, vel_mec_1, freq_lim=1000)
```

```python
vel_mec_1_filt = bandpass_filter(time_mec_1, vel_mec_1,40, 200)
```

```python
plt.figure(figsize = (10,7))
plt.plot(time_mec_1, vel_mec_1_filt)

# Adding labels and title
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Time vs Amplitude')
plt.grid()

# Display the plot
plt.show();
```

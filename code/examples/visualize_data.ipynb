```python
from non_linear_beam.code.module.nonlinearprocess import get_frequency_data, get_time_data, set_file_name
import zipfile
import matplotlib.pyplot as plt
```

```python
dir = "/content/non_linear_beam/Experimental Data/"

time_mec_dir_t1 = set_file_name(dir, time = True, mec = True,test=1)
time_mag_dir_t3 = set_file_name(dir, time = True, mec = False, test=3)

frf_mec_dir_004g = set_file_name(dir, acc = 0.04, time = False, mec = True)
frf_mag_dir_004g = set_file_name(dir,acc = 0.04, time = False, mec = False)
```

```python
time_mec_1, vel_mec_1 = get_time_data(time_mec_dir_t1)
time_mag_3, vel_mag_3 = get_time_data(time_mag_dir_t3)
```

```python
freq_004g_mec, amplitude_004g_mec_sweepup = get_frequency_data(frf_mec_dir_004g)
freq_004g_mag, amplitude_004g_mag_sweepup = get_frequency_data(frf_mag_dir_004g)

_, amplitude_004g_mec_sweepdown = get_frequency_data(frf_mec_dir_004g, sweep_up= False)
_, amplitude_004g_mag_sweepdown= get_frequency_data(frf_mag_dir_004g, sweep_up = False)
```

```python
plt.figure(figsize = (10,7))
plt.plot(freq_004g_mec, amplitude_004g_mec_sweepup, label='Sweep Up', color='blue')
plt.plot(freq_004g_mec, amplitude_004g_mec_sweepdown, label='Sweep Down', color='red')

plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude (m)')
plt.title('Frequency vs Amplitude')
plt.legend()
plt.grid()

plt.show()
```

```python
plt.figure(figsize = (10,7))
plt.plot(time_mec_1, vel_mec_1, label = "Test 1")

# Adding labels and title
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Time vs Amplitude')
plt.grid()
plt.legend()

# Display the plot
plt.show();
```
